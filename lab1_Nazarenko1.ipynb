# Побудуємо довірчі інтервали

```python
#from scipy.stats import norm
#alpha = 1 - 0.95
#norm = norm.ppf(1-alpha/2)
norm = 1.96
```

# Визначення мiнiмального обсягу вибiрки для заданої точностi.

Для абсолютної похибки
n= $\frac{1}{\frac{\epsilon^2}{C*\hat{S}^2*1.96^2}+\frac{1}{N}} $, де С = N^2  або  С = 1, коли оцінка сумарного або середнього відповідно

Для відносної похибки
n= $\frac{1}{\frac{\epsilon^2}{\hat{cv}^2*1.96^2}+\frac{1}{N}}$ $\hat{cv} = \frac{\sqrt{\hat{S}^2}}{\bar{y}} $

```python
#Відносна похибка
def rel(eps, cv, N):
    return 1/((eps/1.96)**2*(1/cv)+1/N)
    #return 1/((eps**2)/((cv**2)*(1.96**2))+1/N)

#Абсолютна похибка
def abss(eps, s, N):
    return 1/((eps/1.96)**2*(1/s)+1/N)

print(rel(0.2, N*np.sqrt(s1)/y_1, N))
print(rel(0.2, np.sqrt(s2)/y_2, N))
print(abss(0.1, s3, N))
```

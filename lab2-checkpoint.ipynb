# 1) Створення вибірки та оцінка параметру

В даній роботі створюю рандомно список із 256 блоків по випадковим 128. Вони можуть повторюватись - це потрібно для подальшого рандомного вибору домогосподарств в відповідних блоках

```python
N = 1024
n = 254
f = 0.25
```

Далі порахуємо значення оцінки відношення $R$. Зробити це можна, похувавши суму TOTINCH поділеного на HHSIZE

TOTINCH -  сумарний загальний дохід

HHSIZE - загальна к-сть людей в домі


```python
table = df1[["BLOCK","UNIT","TOTINCH","HHSIZE"]]
#pd.options.display.max_rows = 20
table
```

```python
R = table["TOTINCH"].sum()/table["HHSIZE"].sum()
R
```

$\hat{R}$ = 18536.527080581243

# 2) Оцінемо параметр трьома методами:

## Лінеаризація Тейлора

```python
s_tot = table["TOTINCH"].var()
s_peop = table["HHSIZE"].var()
s_tot_peop = np.cov(table["TOTINCH"].astype(np.float32), table["HHSIZE"].astype(np.float32))[0][1]#коваріація між TOTINCH та HHSIZE
s_peop_mean = table["HHSIZE"].mean()

#print(s_tot_peop)
#print(table["TOTINCH"].astype(np.float32))
taylor = (1-f)/(s_peop_mean**2*n)*(s_tot-2*R*s_tot_peop+R**2*s_peop)
taylor
```

Значення дисперсії $\hat{D_{Taylor}}(\hat{R})$ = 340356.5125954939 = $583.400816^2$

# Метод Джек-найф

Тут я добавляю в список вибіркове середнє оцінок з вилученням по одному елементу із вибірки. І далі рахую за допомогою функцій
mean() та var() відповідно оцінку відношення та оцінку дисперсії.

```python

R_jk_list = np.zeros((255,))
for i in range(1,len(table["TOTINCH"])):
    R_jk_list[i] = table["TOTINCH"][:-i].sum()/table["HHSIZE"][:-i].sum()
R_jk_list[0] = 25136.84899329 # не було часу вирішити проблеми з 0 в початку списка
R_jk = R_jk_list.mean() #JK оцінка відношення
print("JK оцінка відношення:", R_jk)
R_jk_var = R_jk_list.var() #K оцiнка дисперсiї оцiнки вiдношення
print("JK оцiнка дисперсiї оцiнки вiдношення:", R_jk_var*(n-1)/n)
```

# Побудуємо довірчі інтервали

```python
print(R - 1.96*np.sqrt(taylor), R + 1.96*np.sqrt(taylor)) #довірчий інтервал ліанеразиції Тейлором
print(R - 1.96*np.sqrt(R_jk_var), R + 1.96*np.sqrt(R_jk_var)) #довірчий інтервал методу Джек-найф
```
